```python
from skimage import io
from sklearn.cluster import KMeans
import numpy as np
```

```python
print("Read the image")
image = io.imread('/content/butterflies1.jpg')
io.imshow(image)
io.show()
```

```python
#Dimension of the orginal image
print("Dim of the image")
print("shape of inmage is", image.shape)
print("image is:", type(image))
rows = image.shape[0]
cots = image.shape[1]
print("number of rows is", rows)
print("number of cots is", cots)
```

```python
#flater the image
print("flater the image")
image = image.reshape(rows*cots, 3)
print("shape of flater image is:", image.shape)
print("Flater image is: ", image)
```

```python
#implement K-means clustering form k clusters
print("implement K-means")
kmeans = KMeans(n_clusters = 3)
print(kmeans.fit(image))
```

```python
#replace much pixel
print("compress image")
compressed_image = kmeans.cluster_centers_[kmeans.labels_]
compressed_image = np.clip(compressed_image.astype('uint8'), 0, 255)
```

```python
#reshape the image to original dimmision
print("Reshape the image")
print("compressed image shaper", compressed_image.shape)
compressed_image = compressed_image.reshape(rows, cots, 3)
```

```python
#save and display output image
print("save and display ")
io.imsave("butterflies_compressed_image_3.png", compressed_image)
io.imshow(compressed_image)
io.show
```
